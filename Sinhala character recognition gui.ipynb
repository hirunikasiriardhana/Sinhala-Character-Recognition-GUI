```python
import tkinter as tk
from PIL import ImageTk,Image,ImageDraw#to convert canvas object to image 
import numpy as np
import cv2
import os
import joblib

model=joblib.load('sinhala character knn.sav')


width=500
height=500

font_btn='Helvetica 20 bold'
font_label= 'Helvetica 22 bold'

win=tk.Tk()
count=0

label_dict={0:'අ',1:'එ',2:'ඉ',3:'උ'}

def event_fuction(event):
    x=event.x
    y=event.y
    
    
    x1=x-30
    x2=x+30

    y1=y-30
    y2=y+30

    canvas.create_oval((x1,y1,x2,y2),fill='black')
    img_draw.ellipse((x1,y1,x2,y2),fill='black')
def save():  
    global count
    
    img_array=np.array(img)
    path=os.path.join('data',str(count)+'.jpg')
    
    cv2.imwrite(path,img_array)
    count=count+1

def clear():
    global img,img_draw

    canvas.delete('all')
    img=Image.new('RGB',(width,height),(255,255,255))
    #CREATE NEW IMAGE removing previous image
    img_draw=ImageDraw.Draw(img)
    label_status.config(text='PREDICT CHARACTER:NONE')

def predict():
    img_array=np.array(img)

    img_array=cv2.cvtColor(img_array,cv2.COLOR_BGR2GRAY)#converting to the gray image
    img_array=cv2.resize(img_array,(8,8))#resize to 8x8
    img_array=np.reshape(img_array,(1,64))#reshaping to 1x64 1D

    result=model.predict(img_array)[0]
    #print(label_dict[result])
    label=label_dict[result]
    label_status.config(text='PREDICTED CHARACTER:'+label)
   
   

canvas=tk.Canvas(win,width=width,height=height,bg='white')
canvas.grid(row=0,column=0,columnspan=4)


button_save=tk.Button(win,text='SAVE',bg='green',fg='white',font=font_btn,command=save)
button_save.grid(row=1,column=0)

button_save=tk.Button(win,text='PREDICT',bg='red',fg='white',font=font_btn,command=predict)
button_save.grid(row=1,column=1)

button_save=tk.Button(win,text='CLEAR',bg='pink',fg='white',font=font_btn,command=clear)
button_save.grid(row=1,column=2)

button_save=tk.Button(win,text='EXIT',bg='yellow',fg='white',font=font_btn,command=win.destroy)
button_save.grid(row=1,column=3)

label_status=tk.Label(win,text='PREDICTED CHARACTER: NONE',bg='white',font=font_label)
label_status.grid(row=2,column=0,columnspan=4)

canvas.bind('<B1-Motion>',event_fuction)

#For the image object
img=Image.new('RGB',(width,height),(255,255,255))
img_draw=ImageDraw.Draw(img)#the object to draw in the image 

win.mainloop()
```
